# v_4: Combined Experiment - Average & Plot

## Overview
This notebook processes **all individual training runs** from pruning experiments and creates **aggregated analysis plots**.

### Workflow:
1. **Aggregate**: Load all runs for each (pruning %, batch size), compute mean metrics by batch number
2. **Save**: Write averaged CSV files for each pruning %, batch size combination
3. **Visualize**: Create 6 combined plots per batch size:
   - CE_Train vs Batch Number (all pruning % on one plot)
   - CE_Test vs Batch Number (all pruning % on one plot)
   - Accuracy vs Batch Number (all pruning % on one plot)

### Key Metrics:
- **CE_Train**: Average cross-entropy on training data
- **CE_Test**: Average cross-entropy on test data  
- **Accuracy**: Average test accuracy (%)
- **ln(10)**: Reference line (≈ 2.30) — maximum possible CE for 10-class classification

### Output:
- **CSV files**: `averaged_runs_p_{P%}_bs_{BS}.csv` (one per combination)
- **PNG plots**: `CE_Train_Avg_SLP_MNIST_BS_{BS}.png`, etc.

---

## Cell 1 — Load, Average & Plot

**What this cell does:**

1. **Auto-detects** all pruning percentages from the `prune_layers_ALL/` directory tree (includes fine-grained percentages like 82%, 84%, etc.)
2. **Loads** every individual run `.txt` file for each `(pruning %, batch size)` combination
3. **Averages** `CE_Train`, `CE_Test`, and `Accuracy` across all runs at each batch number
4. **Saves** the averaged result as `averaged_runs_p_{P%}_bs_{BS}.csv` inside the same directory
5. **Plots** three combined figures per batch size:
   - CE Train vs Batch Number
   - CE Test vs Batch Number
   - Accuracy vs Batch Number

**Parameters configured at the top:**
- `BASE_DIR` — root of the `prune_layers_ALL/` tree
- `BATCH_SIZES` — `[64, 1024, 60000]`

**Style:** Consistent with Convolutional-MNIST plots — `font.size=18`, `ylim=(0, 2.7)`, `ln(10)` text marker, reversed legend on CE plots, standard legend on Accuracy.

```python
import pandas as pd
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
import re

# =========================
# CONFIGURATION
# =========================
BASE_DIR = r"C:\Users\Student\Desktop\Neural_research\physlab\SLP\SLP-MNIST\prune_layers_ALL"
BATCH_SIZES = [64, 1024, 60000]
LN10 = np.log(10)

# =========================
# DYNAMICALLY DETECT ALL PRUNING PERCENTAGES
# =========================
p_dirs = glob.glob(os.path.join(BASE_DIR, "p-percentage_*"))
PRUNING_PERCENTAGES = sorted([
    float(re.search(r"p-percentage_([\d.]+)", d).group(1))
    for d in p_dirs
])
print(f"Found {len(PRUNING_PERCENTAGES)} pruning percentages: {PRUNING_PERCENTAGES}\n")

# =========================
# COLOR SCHEME
# =========================
COLOR_LIST_EXTENDED = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#800080",
    "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#B9D9EB", "#17becf",
    "#C5B0D5", "#FDB462", "#80B1D3", "#FB8072", "#BEBADA",
    "#8DD3C7", "#FFFFB3", "#A6D854", "#FB8072", "#FCCDE5"
]
COLOR_LIST = (COLOR_LIST_EXTENDED * ((len(PRUNING_PERCENTAGES) // len(COLOR_LIST_EXTENDED)) + 1))[:len(PRUNING_PERCENTAGES)]
COLOR_LIST = COLOR_LIST[::-1]

# =========================
# STYLE (consistent with Conv-MNIST)
# =========================
plt.rcParams.update({
    "font.size": 18,
    "axes.titlesize": 18,
    "axes.labelsize": 18,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "legend.fontsize": 14,
})

# =========================
# MAIN LOOP: PROCESS EACH BATCH SIZE
# =========================
for bs in BATCH_SIZES:
    print(f"\n{'='*50}\nProcessing batch size: {bs}\n{'='*50}")

    all_avg_dfs = {}

    # STEP 1: Load and average all runs
    for idx, p in enumerate(PRUNING_PERCENTAGES):
        folder = os.path.join(BASE_DIR, f"p-percentage_{p}", f"batch_size_{bs}")
        pattern = os.path.join(folder, f"slp_{p}_{bs}_run_*")
        files = glob.glob(pattern)

        if not files:
            print(f"  [WARNING] No files found for P%={p*100:.1f}, BS={bs}")
            continue

        print(f"  Loading P%={p*100:.1f} ({len(files)} runs)...")

        dfs = []
        skipped = 0
        for f in files:
            try:
                if os.path.getsize(f) == 0:
                    skipped += 1
                    continue
                df = pd.read_csv(f, sep=r"\s+")
                df.columns = df.columns.str.strip()
                df["CE_Train"]    = pd.to_numeric(df["CE_Train"],    errors="coerce")
                df["CE_TEST"]     = pd.to_numeric(df["CE_TEST"],     errors="coerce")
                df["Accuracy(%)"] = pd.to_numeric(df["Accuracy(%)"], errors="coerce")
                dfs.append(df)
            except (pd.errors.EmptyDataError, pd.errors.ParserError):
                skipped += 1

        if skipped > 0:
            print(f"    ⚠ Skipped {skipped} empty/corrupted files")
        if not dfs:
            print(f"  [ERROR] No valid data for P%={p*100:.1f}, BS={bs}")
            continue

        all_runs = pd.concat(dfs, ignore_index=True)
        avg_df = all_runs.groupby("Batch_Number", as_index=False).agg(
            Avg_CE_Train=("CE_Train",    "mean"),
            Avg_CE_Test= ("CE_TEST",     "mean"),
            Avg_Accuracy=("Accuracy(%)", "mean"),
            Num_Runs=    ("CE_TEST",     "count"),
        )

        out_csv = os.path.join(folder, f"averaged_runs_p_{p}_bs_{bs}.csv")
        avg_df.to_csv(out_csv, index=False)
        print(f"    → Saved: {os.path.basename(out_csv)}")
        all_avg_dfs[p] = avg_df

    if not all_avg_dfs:
        print(f"  [SKIP] No data for BS={bs}")
        continue

    # =========================
    # PLOT HELPER
    # =========================
    def make_plot(y_col, ylabel, ylim, subtitle, filename, reverse_legend=True, legend_loc="upper right"):
        plt.figure(figsize=(12, 6))
        ax = plt.gca()

        for idx, (p, avg_df) in enumerate(all_avg_dfs.items()):
            color = COLOR_LIST[idx % len(COLOR_LIST)]
            ax.plot(avg_df["Batch_Number"], avg_df[y_col],
                    label=f"P%={p*100:.1f}", color=color)

        ax.set_xlabel("Batch Number")
        ax.set_ylabel(ylabel)
        ax.set_ylim(*ylim)

        if "CE" in y_col:
            ax.text(0.06, LN10, r"$\ln(10)$",
                    transform=ax.get_yaxis_transform(),
                    fontsize=14, va="center")

        ax.text(0.5, 0.95, "Cross-Entropy" if "CE" in y_col else "Accuracy",
                transform=ax.transAxes, ha="center", va="top")
        ax.text(0.5, 0.90, f"({subtitle}, BS={bs})",
                transform=ax.transAxes, ha="center", va="top", fontsize=16)

        handles, labels = ax.get_legend_handles_labels()
        if reverse_legend:
            ax.legend(handles[::-1], labels[::-1], frameon=False, loc=legend_loc)
        else:
            ax.legend(handles, labels, frameon=False, loc=legend_loc)

        ax.grid(True)
        plt.tight_layout()
        plt.savefig(filename, dpi=300, bbox_inches="tight")
        print(f"  → Saved: {os.path.basename(filename)}")
        plt.show()

    make_plot("Avg_CE_Train", "Average CE", (0, 2.7),
              "SLP Average of MNIST Training-Vectors",
              f"CE_Train_Avg_SLP_MNIST_BS_{bs}.png")

    make_plot("Avg_CE_Test", "Average CE", (0, 2.7),
              "SLP Average of MNIST Test-Vectors",
              f"CE_Test_Avg_SLP_MNIST_BS_{bs}.png")

    make_plot("Avg_Accuracy", "Average Accuracy (%)", (0, 100),
              "SLP Average of MNIST Test-Vectors",
              f"Accuracy_Avg_SLP_MNIST_BS_{bs}.png",
              reverse_legend=False, legend_loc="lower right")

print(f"\n{'='*50}\n✓ All plots completed successfully!\n{'='*50}")
```

## Summary of Outputs

### CSV Files Generated
For each batch size (BS = 64, 1024, 60000) and pruning percentage (P% = 0%, 10%, ..., 100%):
- **File**: `prune_layers_ALL/p-percentage_{P%}/batch_size_{BS}/averaged_runs_p_{P%}_bs_{BS}.csv`
- **Columns**: `Batch_Number | Avg_CE_Train | Avg_CE_Test | Avg_Accuracy | Num_Runs`
- **Interpretation**: Each row is the average across ~100 stochastic runs at that batch number

### PNG Plots Generated
For each batch size:

2. **CE_Test_Avg_SLP_MNIST_BS_{BS}.png** — Test CE vs batch number
3. **Accuracy_Avg_SLP_MNIST_BS_{BS}.png** — Test accuracy vs batch number

Each plot shows 11 lines (one per P% = 0%, 10%, ..., 100%) in distinct colors.

### Key Features
- **ln(10) reference line** on CE plots marks the maximum possible CE for 10-class classification
- **Reversed legend** on CE plots (P%=100% at top) for visual clarity of learning progression
- **Standard legend** on Accuracy plots (P%=100% at bottom)
- **High-resolution output** (300 DPI) suitable for publication

- Identify optimal pruning % for this dataset and architecture
