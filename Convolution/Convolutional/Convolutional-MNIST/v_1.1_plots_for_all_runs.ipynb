```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

# =========================
# CONFIG (Convolutional MNIST)
# =========================
BASE_DIR_ROOT = r"C:\Users\Student\Desktop\Neural_research\physlab\Convolution\Convolutional\Convolutional-MNIST"

PRUNE_LAYERS_OPTIONS = ['CONV', 'FHL', 'SHL', 'FHL+SHL', 'ALL']

PRUNE_LAYER_DIR_MAP = {
    "CONV": "prune_layers_CONV",
    "FHL": "prune_layers_FHL",
    "SHL": "prune_layers_SHL",
    "FHL+SHL": "prune_layers_FHL+SHL",
    "ALL": "prune_layers_ALL"
}

BATCH_DIR_TEMPLATE = "p-percentage_{:.1f}\\batch_size_{}"
FILE_PATTERN = "convol_{:.1f}_{}_run_*.txt"

BATCH_SIZES = [1024]
PRUNING_LEVELS = [round(x * 0.1, 1) for x in range(11)]

# =========================
# OUTPUT DIRECTORY
# =========================
PLOT_ROOT = os.path.join(BASE_DIR_ROOT, "Conv_raw_plot_1024")
os.makedirs(PLOT_ROOT, exist_ok=True)

# =========================
# STYLE
# =========================
plt.rcParams.update({
    "font.size": 18,
    "axes.labelsize": 18,
    "xtick.labelsize": 18,
    "ytick.labelsize": 18
})

# =========================
# PLOTTING FUNCTION
# =========================
def plot_all_runs(metric_col, ylabel, metric_name, pruning_p,
                  batch_size, files, prune_layer, plot_dir):

    plt.figure(figsize=(12, 6))

    for run_file in files:
        df = pd.read_csv(run_file, sep=r"\s+")
        df.columns = df.columns.str.strip()
        df[metric_col] = pd.to_numeric(df[metric_col], errors="coerce")
        plt.plot(df["Batch_Number"], df[metric_col], alpha=0.5)

    # Axis limits
    if metric_name in ["ce_train", "ce_test"]:
        plt.ylim(0, 2.7)
    else:
        plt.ylim(0, 100)

    plt.xlabel("Batch Number")
    plt.ylabel(ylabel)
    plt.grid(True)

    pruning_str = f"P% = {int(pruning_p * 100)}%"

    # =========================
    # TITLES (FIXED POSITIONS)
    # =========================
    if metric_name == "ce_train":

        plt.text(0.5, 0.97, f"Convolutional MNIST ({prune_layer}-layers)",
             transform=plt.gca().transAxes, ha='center', va='top')
        plt.text(0.5, 0.93,
                 f"(Training-Vectors, {pruning_str}, BS={batch_size})",
                 transform=plt.gca().transAxes, ha='center', va='top', fontsize=16)

    elif metric_name == "ce_test":

        plt.text(0.5, 0.97, f"Convolutional MNIST ({prune_layer}-layers)",
             transform=plt.gca().transAxes, ha='center', va='top')
    
        plt.text(0.5, 0.93,
                 f"(Test-Vectors, {pruning_str}, BS={batch_size})",
                 transform=plt.gca().transAxes, ha='center', va='top', fontsize=16)

    else:  # accuracy

        plt.text(0.5, 0.55, f"Convolutional MNIST ({prune_layer}-layers)",
                 transform=plt.gca().transAxes, ha='center', va='top')
        plt.text(0.5, 0.50,
                 f"(Test-Vectors, {pruning_str}, BS={batch_size})",
                 transform=plt.gca().transAxes, ha='center', va='top', fontsize=16)

    plt.tight_layout()

    filename = f"Conv_{prune_layer}_{metric_name}_P{int(pruning_p*100)}_BS{batch_size}.png"
    plt.savefig(os.path.join(plot_dir, filename), dpi=300, bbox_inches="tight")
    plt.close()

    print(f"[SAVED] {filename}")

# =========================
# MAIN LOOP
# =========================
for prune_layer in PRUNE_LAYERS_OPTIONS:

    plot_dir = os.path.join(PLOT_ROOT, prune_layer)
    os.makedirs(plot_dir, exist_ok=True)

    base_dir = os.path.join(BASE_DIR_ROOT, PRUNE_LAYER_DIR_MAP[prune_layer])

    for batch_size in BATCH_SIZES:
        for p in PRUNING_LEVELS:

            folder = os.path.join(base_dir, BATCH_DIR_TEMPLATE.format(p, batch_size))
            files = glob.glob(os.path.join(folder, FILE_PATTERN.format(p, batch_size)))

            if not files:
                continue

            plot_all_runs("CE_Train", "Cross-Entropy", "ce_train",
                          p, batch_size, files, prune_layer, plot_dir)

            plot_all_runs("CE_TEST", "Cross-Entropy", "ce_test",
                          p, batch_size, files, prune_layer, plot_dir)

            plot_all_runs("Accuracy(%)", "Accuracy (%)", "accuracy",
                          p, batch_size, files, prune_layer, plot_dir)
```

```python
# THIS PLOT IS FOR PLOTTING USING RUNNING AVERAGE



import pandas as pd
import glob
import os
import matplotlib.pyplot as plt

# =========================
# CONFIG (Convolutional MNIST)
# =========================
BASE_DIR_ROOT = r"C:\Users\Student\Desktop\Neural_research\physlab\Convolution\Convolutional\Convolutional-MNIST"

PRUNE_LAYERS_OPTIONS = ['CONV', 'FHL', 'SHL', 'FHL+SHL', 'ALL']

PRUNE_LAYER_DIR_MAP = {
    "CONV": "prune_layers_CONV",
    "FHL": "prune_layers_FHL",
    "SHL": "prune_layers_SHL",
    "FHL+SHL": "prune_layers_FHL+SHL",
    "ALL": "prune_layers_ALL"
}

BATCH_DIR_TEMPLATE = "p-percentage_{:.1f}\\batch_size_{}"
FILE_PATTERN = "convol_{:.1f}_{}_run_*.txt"

BATCH_SIZES = [64, 1024]
PRUNING_LEVELS = [round(x * 0.1, 1) for x in range(11)]

RUNNING_WINDOW = 10 # ⭐ running average window

# =========================
# OUTPUT DIRECTORY (RENAMED)
# =========================
PLOT_ROOT = os.path.join(BASE_DIR_ROOT, "Conv_running_average")
os.makedirs(PLOT_ROOT, exist_ok=True)

# =========================
# STYLE
# =========================
plt.rcParams.update({
    "font.size": 18,
    "axes.labelsize": 18,
    "xtick.labelsize": 18,
    "ytick.labelsize": 18
})

# =========================
# PLOTTING FUNCTION
# =========================
def plot_all_runs(metric_col, ylabel, metric_name, pruning_p,
                  batch_size, files, prune_layer, plot_dir):

    plt.figure(figsize=(12, 6))

    for run_file in files:
        df = pd.read_csv(run_file, sep=r"\s+")
        df.columns = df.columns.str.strip()

        df[metric_col] = pd.to_numeric(df[metric_col], errors="coerce")

        # =========================
        # RUNNING AVERAGE (window=50)
        # =========================
        smoothed = (
            df[metric_col]
            .rolling(window=RUNNING_WINDOW, min_periods=1)
            .mean()
        )

        plt.plot(df["Batch_Number"], smoothed, alpha=0.5)

    # Axis limits
    if metric_name in ["ce_train", "ce_test"]:
        plt.ylim(0, 2.7)
    else:
        plt.ylim(0, 100)

    plt.xlabel("Batch Number")
    plt.ylabel(ylabel)
    plt.grid(True)

    pruning_str = f"P% = {int(pruning_p * 100)}%"

    # =========================
    # TITLES (FIXED POSITIONS)
    # =========================
    if metric_name == "ce_train":

        plt.text(
            0.5, 0.97,
            f"Convolutional MNIST ({prune_layer}-layers)",
            transform=plt.gca().transAxes,
            ha='center', va='top'
        )

        plt.text(
            0.5, 0.93,
            f"(Training-Vectors, {pruning_str}, BS={batch_size})",
            transform=plt.gca().transAxes,
            ha='center', va='top', fontsize=16
        )

    elif metric_name == "ce_test":

        plt.text(
            0.5, 0.97,
            f"Convolutional MNIST ({prune_layer}-layers)",
            transform=plt.gca().transAxes,
            ha='center', va='top'
        )

        plt.text(
            0.5, 0.93,
            f"(Test-Vectors, {pruning_str}, BS={batch_size})",
            transform=plt.gca().transAxes,
            ha='center', va='top', fontsize=16
        )

    else:  # accuracy

        plt.text(
            0.5, 0.55,
            f"Convolutional MNIST ({prune_layer}-layers)",
            transform=plt.gca().transAxes,
            ha='center', va='top'
        )

        plt.text(
            0.5, 0.50,
            f"(Test-Vectors, {pruning_str}, BS={batch_size})",
            transform=plt.gca().transAxes,
            ha='center', va='top', fontsize=16
        )

    plt.tight_layout()

    filename = f"Conv_{prune_layer}_{metric_name}_P{int(pruning_p*100)}_BS{batch_size}.png"
    plt.savefig(os.path.join(plot_dir, filename), dpi=300, bbox_inches="tight")
    plt.close()

    print(f"[SAVED] {filename}")

# =========================
# MAIN LOOP
# =========================
for prune_layer in PRUNE_LAYERS_OPTIONS:

    plot_dir = os.path.join(PLOT_ROOT, prune_layer)
    os.makedirs(plot_dir, exist_ok=True)

    base_dir = os.path.join(BASE_DIR_ROOT, PRUNE_LAYER_DIR_MAP[prune_layer])

    for batch_size in BATCH_SIZES:
        for p in PRUNING_LEVELS:

            folder = os.path.join(base_dir, BATCH_DIR_TEMPLATE.format(p, batch_size))
            files = glob.glob(os.path.join(folder, FILE_PATTERN.format(p, batch_size)))

            if not files:
                continue

            plot_all_runs(
                "CE_Train", "Cross-Entropy", "ce_train",
                p, batch_size, files, prune_layer, plot_dir
            )

            plot_all_runs(
                "CE_TEST", "Cross-Entropy", "ce_test",
                p, batch_size, files, prune_layer, plot_dir
            )

            plot_all_runs(
                "Accuracy(%)", "Accuracy (%)", "accuracy",
                p, batch_size, files, prune_layer, plot_dir
            )
```
